```python
import tensorflow as tf
print(tf.__version__)
```

```python
train_X = []
train_Y = []
for user in users:
    data = []
    data.append(user.birth.year)
    y_data = []
    if user.gender:
        data.append(1)
    else:
        data.append(0)

    for tag in tags:
        if tag in user.like_tags.all():
            data.append(1)
        else:
            data.append(0)
    for newsletter in newsletters:
        if newsletter in user.subscribes.all():
            data.append(1)
        else:
            data.append(0)
    tensor_length = len(data)
    train_X.append(data)

print(train_X)
print(tensor_length)
```

```python
import random
```

```python
for i in range(1000):
    if i % 100 == 0:
        print(i)
    data = []
    data.append(random.randint(1980,2001))
    data.append(random.randint(0,1))
    for tag in tags:
        data.append(random.randint(0,1))
    for newsletter in newsletters:
        data.append(random.randint(0,1))
    train_X.append(data)

print(train_X)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=32, input_shape=(tensor_length, ), activation='relu'),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=152, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(train_X, train_Y, validation_split=0.25)
```
